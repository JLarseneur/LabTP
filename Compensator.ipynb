```python
df_comp = pd.read_excel("fullMeasure_modif.xlsx", sheet_name="Feuil1")

df_res = df_comp[["Name", " crossIncl", " lengthIncl", ' dHz', 'face', 'cycle']]
df_res.loc[:, " crossIncl"] = df_res.loc[:, " crossIncl"] + np.pi/2
df_res.loc[:, " lengthIncl"] = df_res.loc[:, " lengthIncl"] + np.pi/2
df_res.loc[:, ' dHz_cross'] = df_res.loc[:, ' dHz'] + np.pi/2

df_length = pd.DataFrame(np.array([np.sin(df_res[' lengthIncl']) * np.cos(df_res[' dHz']),
                                   np.sin(df_res[' lengthIncl']) * np.sin(df_res[' dHz']),
                                   np.cos(df_res[' lengthIncl'])]).T,
                         columns=["length_vect_x", "length_vect_y", "length_vect_z"])
df_cross = pd.DataFrame(np.array([np.sin(df_res[' crossIncl']) * np.cos(df_res[' dHz_cross']),
                                 np.sin(df_res[' crossIncl']) * np.sin(df_res[' dHz_cross']),
                                 np.cos(df_res[' crossIncl'])]).T,
                       columns=["cross_vect_x", "cross_vect_y", "cross_vect_z"])
df_res = pd.concat([df_res, df_length, df_cross], axis=1, sort=False)

df_cross_prod = df_res.apply(lambda row: np.cross(np.array([row["length_vect_x"], row["length_vect_y"], row["length_vect_z"]]),\
                                                  np.array([row["cross_vect_x"], row["cross_vect_y"], row["cross_vect_z"]])),\
                             axis=1)
df_res["cross_prod_x"] = df_cross_prod.apply(lambda row: row[0])
df_res["cross_prod_y"] = df_cross_prod.apply(lambda row: row[1])
df_res["cross_prod_z"] = df_cross_prod.apply(lambda row: row[2])
df_res.loc[:, "great_slope"] = np.arctan(np.sqrt(np.power(df_res["cross_prod_x"], 2) + np.power(df_res["cross_prod_y"], 2)) /\
                                  df_res["cross_prod_z"]) / np.pi * 200

df_res["great_slope"].plot()
plt.show()

## Stats par cycles (pour tudier la variabilit en fonction de la plus grande pente)
## et par prismes (n'a pas vraiment de sens au vu de cohrence des rsultats)
```
